```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('BostonHousing.csv')

# Display basic information
print("Dataset Shape:", df.shape)
print("\nFirst 5 Rows:")
print(df.head())
print("\nData Types and Non-Null Counts:")
print(df.info())
print("\nSummary Statistics:")
print(df.describe())

# Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())

# Visualizations
plt.figure(figsize=(12, 8))

# Histograms for all numerical columns
df.hist(bins=20, figsize=(15, 10))
plt.tight_layout()
plt.show()

# Correlation matrix heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()

# Boxplots for all numerical columns
plt.figure(figsize=(12, 8))
df.boxplot(figsize=(15, 8))
plt.xticks(rotation=45)
plt.title("Boxplots of Numerical Features")
plt.show()

# Scatter plots for selected features against target (medv)
plt.figure(figsize=(12, 8))
sns.scatterplot(x='rm', y='medv', data=df)
plt.title("Room vs. Median Home Value")
plt.show()

plt.figure(figsize=(12, 8))
sns.scatterplot(x='lstat', y='medv', data=df)
plt.title("Lower Status Population % vs. Median Home Value")
plt.show()

# Pairplot for selected features
sns.pairplot(df[['crim', 'rm', 'age', 'dis', 'tax', 'ptratio', 'lstat', 'medv']])
plt.show()
```
