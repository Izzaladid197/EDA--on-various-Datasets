```python
# 📌 Step 1: Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 🔼 Optional: Set plot style
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

# 📁 Step 2: Load dataset
df = pd.read_csv('/content/Mall_Customers.csv')

# 🧼 Step 3: Clean column names
df.columns = df.columns.str.strip()

# 👁️ Step 4: Preview data
print("First 5 rows:\n", df.head())

# 🔎 Step 5: Dataset info
print("\nInfo:")
print(df.info())

print("\nSummary statistics:")
print(df.describe())

# 🧪 Step 6: Check for missing or duplicate data
print("\nMissing values:\n", df.isnull().sum())
print("\nDuplicate rows:", df.duplicated().sum())

# 📊 Step 7: Categorical analysis
if 'Gender' in df.columns:
    print("\nGender distribution:\n", df['Gender'].value_counts())
    sns.countplot(x='Gender', data=df)
    plt.title('Gender Distribution')
    plt.show()

# 📈 Step 8: Numerical feature distributions
num_cols = ['Age', 'Annual Income (k$)', 'Spending Score (1-100)']
for col in num_cols:
    sns.histplot(df[col], kde=True, bins=15)
    plt.title(f'{col} Distribution')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()

# 📦 Step 9: Box plots for outlier detection
for col in ['Annual Income (k$)', 'Spending Score (1-100)']:
    sns.boxplot(y=col, data=df)
    plt.title(f'Boxplot of {col}')
    plt.show()

# 🔗 Step 10: Correlation heatmap
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()

# 🧬 Step 11: Pairplot with hue as Gender (if exists)
if 'Gender' in df.columns:
    sns.pairplot(df.drop(columns=['CustomerID']), hue='Gender')
    plt.show()
```
