```python
# ðŸ“¦ Step 1: Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# ðŸ“ Step 2: Load the dataset
df = pd.read_csv('/content/cluster_moons.csv')

# ðŸ§¼ Step 3: Clean column names
df.columns = df.columns.str.strip()

# ðŸ‘€ Step 4: View basic structure
print("ðŸ”¹ First 5 rows:\n", df.head())
print("\nðŸ”¹ Data Info:")
print(df.info())
print("\nðŸ”¹ Summary Statistics:")
print(df.describe())

# â“ Step 5: Check for missing values and duplicates
print("\nðŸ”¹ Missing Values:\n", df.isnull().sum())
print("\nðŸ”¹ Duplicate Rows:", df.duplicated().sum())

# ðŸ“Š Step 6: Distribution plots
for col in df.select_dtypes(include=[np.number]).columns:
    sns.histplot(df[col], kde=True, bins=20)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()

# ðŸ§° Step 7: Scatter plot if 2D data
if df.shape[1] == 2:
    plt.scatter(df.iloc[:, 0], df.iloc[:, 1], alpha=0.7)
    plt.xlabel(df.columns[0])
    plt.ylabel(df.columns[1])
    plt.title("2D Scatter Plot of Features")
    plt.grid(True)
    plt.show()

# ðŸ”— Step 8: Correlation heatmap
if df.select_dtypes(include=[np.number]).shape[1] > 1:
    sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', linewidths=0.5)
    plt.title('Correlation Heatmap')
    plt.show()

# ðŸŒ Step 9: Pairplot (if number of columns is reasonable)
if df.select_dtypes(include=[np.number]).shape[1] <= 5:
    sns.pairplot(df)
    plt.show()
```
